```python
class Complaint:
    def __init__(self, user, title, description):
        self.user = user
        self.title = title
        self.description = description
        self.resolved = False

    def resolve(self):
        self.resolved = True

    def __str__(self):
        status = "Resolved" if self.resolved else "Pending"
        return f"[{status}] {self.user} - {self.title} - {self.description}"


class ComplaintManagementSystem:
    def __init__(self):
        self.complaints = []

    def register_complaint(self, user, title, description):
        complaint = Complaint(user, title, description)
        self.complaints.append(complaint)

    def view_complaints(self):
        for idx, complaint in enumerate(self.complaints, 1):
            print(f"{idx}. {complaint}")

    def resolve_complaint(self, idx):
        self.complaints[idx - 1].resolve()


def main():
    cms = ComplaintManagementSystem()

    while True:
        print("\n--- Complaint Management System ---")
        print("1. Register a complaint")
        print("2. View all complaints")
        print("3. Resolve a complaint")
        print("4. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            user = input("Enter your name: ")
            title = input("Enter complaint title: ")
            description = input("Enter complaint description: ")
            cms.register_complaint(user, title, description)

        elif choice == 2:
            cms.view_complaints()

        elif choice == 3:
            cms.view_complaints()
            idx = int(input("Enter the number of the complaint to resolve: "))
            cms.resolve_complaint(idx)

        elif choice == 4:
            print("Exiting...")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```
